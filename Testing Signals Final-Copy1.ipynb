```python
###################################################
"""Feature Extraction"""
###################################################
def MAV(data):
    return np.mean(np.abs(data))

def WL(data):
    WL = 0
    for i in range(0,len(data)):
        if i+1 < len(data):
            WL_i = data.iloc[i+1]-data.iloc[i]
            WL = WL + abs(WL_i)
    return WL

Features = Subjects_data.agg([MAV])
#Features = Subjects_data.agg([WL])
#Features = Subjects_data.agg([MAV, WL])
Features = Features.reset_index()
```

```python
scaler = StandardScaler()
model = SVC()

accuracy = []
f1score = []
model_coef = []
for i in range(1,72):
    DataTrain = Features[Features['Subject']!=i]
    DataTest = Features[Features['Subject']==i]
    
    X_train = DataTrain.iloc[:,2:].values
    y_train = DataTrain['class'].values
    
    X_test = DataTest.iloc[:,2:].values
    y_test = DataTest['class'].values

    scaler.fit(X_train)
    X_train_norm = scaler.transform(X_train)
    X_test_norm = scaler.transform(X_test)
    
    model.fit(X_train_norm, y_train)
    #model_coef.append(model.coef_[0])
    
    y_pred = model.predict(X_test_norm)
    accuracy.append(accuracy_score(y_test, y_pred))
    f1score.append(f1_score(y_test, y_pred, average='macro'))
    
print(f'Average accuracy = {np.mean(np.array(accuracy))*100} %')
```
